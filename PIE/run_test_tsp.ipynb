```python
import pickle
import time
import numpy as np
name = "TSP_hard"
with open(f'dataset/{name}.pkl', 'rb') as file:
    data = pickle.load(file)
```

```python
def read_graph(g):
    node_number = g.number_of_nodes()
    G = [[0 for _ in range(node_number)] for _ in range(node_number)]
    for edge in g.edges(data=True):
        G[edge[0]][edge[1]] = edge[2]['weight']
        G[edge[1]][edge[0]] = edge[2]['weight']
    return G
```

```python
## deepseek
def travelling_salesman(G):
    n = len(G)
    ans = float('inf')
    
    for start in range(n):
        visited = [False] * n
        visited[start] = True
        path_length = 0
        current = start
        
        for _ in range(n - 1):
            next_node = -1
            min_dist = float('inf')
            
            for i in range(n):
                if not visited[i] and G[current][i] < min_dist:
                    min_dist = G[current][i]
                    next_node = i
            
            path_length += min_dist
            visited[next_node] = True
            current = next_node
        
        path_length += G[current][start]
        ans = min(ans, path_length)
    
    return ans
```

```python
tot = len(data)
correct = 0
error = 0
feasible = 0
node, edge = [], []

t1 = time.time()
for i, x in enumerate(data):
    exact_answer, graph = x["exact_answer"], x["graph"]
    node_number, edge_number = graph.number_of_nodes(), graph.number_of_edges()
    node.append(node_number)
    edge.append(edge_number)
    G = read_graph(graph)
    predicted_answer = travelling_salesman(G)
    if (exact_answer != predicted_answer):
        print(f"{i}th result is wrong")
        print(f"exact_answer is {exact_answer}")
        print(f"predicted_answer is {predicted_answer}")
    else:
        correct += 1
    error += (abs(predicted_answer - exact_answer) / exact_answer)
    if (predicted_answer >= exact_answer):
        feasible += 1
print(f"correct rate is {correct / tot * 100:.2f}%")
print(f"approximation ratio is {error / tot:.5f}")
print(f"feasible rate is {feasible / tot * 100:.2f}%")
t2 = time.time()
print(f"cost time is {t2 - t1:.2f}s")    
print(f"node number: {int(np.mean(node))}")
print(f"edge number: {int(np.mean(edge))}")
```
