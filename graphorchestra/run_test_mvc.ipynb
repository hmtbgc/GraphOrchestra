```python
import pickle
from tqdm import tqdm
import numpy as np

name = "MVC_hard"
with open(f'new_dataset/{name}.pkl', 'rb') as file:
    data = pickle.load(file)
```

```python
## deepseek
def minimum_vertex_cover(G):
    vertex_cover = set()
    edges = set()
    
    for vertex in range(len(G)):
        for neighbor in G[vertex]:
            if (vertex, neighbor) not in edges and (neighbor, vertex) not in edges:
                edges.add((vertex, neighbor))
    
    while edges:
        max_coverage = -1
        best_vertex = -1
        
        for vertex in range(len(G)):
            covered_edges = {edge for edge in edges if vertex in edge}
            coverage = len(covered_edges)
            if coverage > max_coverage:
                max_coverage = coverage
                best_vertex = vertex
        
        vertex_cover.add(best_vertex)
        edges = {edge for edge in edges if best_vertex not in edge}
    
    return len(vertex_cover)
```

```python
def read_graph(g):
    node_number = g.number_of_nodes()
    G = [[] for _ in range(node_number)]
    for edge in g.edges():
        # print("edge == ", edge)
        G[edge[0]].append(edge[1])
    return G
```

```python
import time
```

```python
tot = len(data)
print(tot)
correct = 0
error = 0
feasible = 0
t1 = time.time()
node, edge = [], []
for i, x in tqdm(enumerate(data)):
    exact_answer, graph = x["exact_answer"], x["graph"]
    node.append(graph.number_of_nodes())
    edge.append(graph.number_of_edges())
    G = read_graph(graph)
    predicted_answer = minimum_vertex_cover(G)
    if (exact_answer != predicted_answer):
        print(f"{i}th result is wrong")
        print(f"exact_answer is {exact_answer}")
        print(f"predicted_answer is {predicted_answer}")
    else:
        correct += 1
    error += (abs(predicted_answer - exact_answer) / exact_answer)
    if (predicted_answer >= exact_answer):
        feasible += 1
print(f"correct rate is {correct / tot * 100:.2f}%")
print(f"approximation ratio is {error / tot:.5f}")
print(f"feasible rate is {feasible / tot * 100:.2f}%")
t2 = time.time()
print(f"tot time is {t2 - t1:.2f}s")
print(f"node: {int(np.mean(node))}")
print(f"edge: {int(np.mean(edge))}")
```
